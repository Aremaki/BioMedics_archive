```python
import pandas as pd

SAVE_DIR = "/export/home/cse200093/Jacques_Bio/data_bio/snomed_loinc_mapping/loinc_snomed_mapping.csv"
```

```python
# Import ohdsi tables of interest
# Sources :
# - https://athena.ohdsi.org/vocabulary/list
# - https://www.ohdsi.org/wp-content/uploads/2020/10/Polina-Talapova-Talapova-LOINC-SNOMED-crosswalk-full-paper-updated.pdf
# - https://github.com/OHDSI/Vocabulary-v5.0/tree/b01501774d94067574cd7ff18d1c5baae01c7881

concept_relationship_df = pd.read_csv('/export/home/cse200093/Jacques_Bio/data_bio/snomed_loinc_mapping/CONCEPT_RELATIONSHIP.csv', sep="\t")
concept_df = pd.read_csv("/export/home/cse200093/Jacques_Bio/data_bio/snomed_loinc_mapping/CONCEPT.csv", sep="\t")
```

```python
# Show them

print("Concept relationship:")
display(concept_relationship_df)
print("Concept:")
display(concept_df)
```

```python
# Retrieve SNOMED concepts

snomed_df = concept_df[concept_df["vocabulary_id"] == "SNOMED"][["concept_id", "concept_name", "concept_code", "concept_class_id"]]
snomed_df = snomed_df.rename(columns={"concept_id":"snomed_concept_id", "concept_name":"snomed_term", "concept_code":"snomed_code", "concept_class_id":"snomed_class_id"})
snomed_df
```

```python
# Retrieve LOINC concepts

loinc_df = concept_df[concept_df["vocabulary_id"] == "LOINC"][["concept_id", "concept_name", "concept_code", "concept_class_id"]]
loinc_df = loinc_df.rename(columns={"concept_id":"loinc_concept_id", "concept_name":"loinc_term", "concept_code":"loinc_code", "concept_class_id":"loinc_class_id"})
loinc_df
```

```python
# Link SNOMED and loincs

snomed_loinc_link = concept_relationship_df[["concept_id_1", "concept_id_2"]].rename(columns={"concept_id_1":"loinc_concept_id", "concept_id_2":"snomed_concept_id"})
final_df = snomed_loinc_link.merge(loinc_df, on="loinc_concept_id", how="inner")
final_df = final_df.merge(snomed_df, on="snomed_concept_id", how="inner").drop(columns=["loinc_concept_id", "snomed_concept_id"])
final_df
```

```python
final_df.to_csv(SAVE_DIR)
```
